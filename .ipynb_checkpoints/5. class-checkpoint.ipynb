```python
# 생성자, 소멸자 self(this)
# 생성자 - 객체 생성시 자동 호출, 객체 초기화에 사용된다.
# 소멸자 - 객체 소멸시 자동 호출, 메모리 정리에 사용된다.
 # Java는 소멸자가 없다.
    
class User:
    name=""
    age=0
    address=""
    
#     def __init__(self): # default 생성자 - 생성자를 구현하면 사라진다.
#         print("생성자")
#         # pass # 맴버 변수를 자동초기화 한다.
    # 파이썬은 오버로드가 없기 때문에 초기값을 사용한다.
    def __init__(self, name="", age=0, address=""): # default 생성자 - 생성자를 구현하면 사라진다.
        print("생성자2")
        self.name=name
        self.age=age
        self.address=address
    
    def __del__(self): # 소멸자 - 객체를 지울대 호출한다. - 메모리를 정리한다.
        print("소멸자")
        
    def setName(self,name):
        self.name=name
    def setAge(self, age):
        self.age=age
    def setAddress(self,address):
        self.address=address
        
    def getName(self):
        return self.name
    def getAge(self):
        return self.age
    def getAddress(self):
        return self.address
    
kim=User()

kim.setName("김유신")
kim.setAge(30)
kim.setAddress("서울시")
print("이름: " +kim.getName())
print("이름: " +str(kim.getAge()))
print("이름: " +kim.getAddress())
print()

lee=User("siri",29,"수원")

print("이름: " +lee.getName())
print("이름: " +str(lee.getAge()))
print("이름: " +lee.getAddress())
print()

hong=User("홍길동",200)
print("이름: " +hong.getName())
print("이름: " +str(hong.getAge()))
print("이름: " +hong.getAddress())
print()

del kim # 데이터 삭제하기
```

```python
# 연산자 멤버 함수/ opreator 함수
class Command :
    str=""
    def __init__(self,str):
        self.str=str
    def setStr( self, str):
        self.str=str
    def getStr( self):
        return self.str
    def __add__( self, other ): # +
        return self.str+other.str 
    def __mul__( self, other):  # *
        return self.str *2 + other.str *2 
    def __and__( self, other): # &
        return list(zip( self.str, other.str))

# 문자열 +문자열은 가능하지만 Comand끼리는 더할수 없다.
a= Command("ABCD")
b= Command("EFGH")
print(a+b) # 클래스로 연산자 맴버 함수를 지정해줘야 한다.
print(a*b) #객체의 시작주소*시작주소라 답이 나오지 않는다.
print(a&b)
print("abcd"+"efgh")
```

```python
class Pet:
    name=""
    def __init__(self, name=""):
        self.name = name
    def getName(self):
        return self.name
    
class Dog(Pet):
    sound=""
    def __init__(self, name="", sound=""):
        Pet.__init__(self, name)
        self.sound = sound
        
    
    def getSound(self):
        return self.sound
        
class Pug(Dog):
    size=0
    def __init__(self, name="", sound="", size=0):
        Dog.__init__(self, name, sound)
        self.size=size
        
    def getSize(self):
        return self.size
    
class Kitty(Pug):
    color=""
    # 1번째 방법
#     def __init__(self, name="", sound="", size=0, color=""):
#         Pug.__init__(self,name,sound,size)
#         self.color=color
    # 2번째 방법!
    def __init__(self, name="", sound="", color=""):
        Dog.__init__(self, name, sound)
        self.color=color
        
    def getColor(self):
        return self.color
    
pug=Pug("퍼그", "멍멍", 20)
print("이름: "+pug.getName())
print("소리: "+pug.getSound())
print("크기: "+ str(pug.getSize()))

kitty=Kitty("키티","야옹야옹","분홍색")
print("이름 : "+kitty.getName())
print("소리 : "+kitty.getSound())
print("색상 : "+kitty.getColor())
```

```python
# 오버라이드 부모 클래스의 함수가 자식 클래스에도 존재한다.

class Pet:
    name=""
    def __init__(self, name=""):
        self.name = name
    def disp(self):
        print("이름 : "+self.name)
    
class Dog(Pet) :
    sound=""
    def __init__(self,name="", sound=""):
        Pet.__init__(self,name)
        self.sound=sound
    def disp(self):
        print("소리: "+self.sound)
        
class Pug(Dog):
    size=0
    def __init__(self, name="",sound="",size=0):
        Dog.__init__(self,name,sound)
        self.size=size
    def disp(self): # 오버라이드 함수(재정의)
        print("크기: "+str(self.size))
    
pug=Pug("해피", "왈왈왈",30)
pug.disp()

pet=Pet("강아지")
pet.disp()

dog=Dog( "개", "멍멍멍" )
dog.disp()
```

```python
# 다중 상속 - 부모 클래스가 두개 이상이다.
# c - 다중 상속이 가능 / 사용 불가능
# Java - 다중상속 구현 불가능 / 사용 불가능, 인터페이스 대체
# Python - 다중 상속 구현 가능 / 사용 가능
class Animal: # 부모클래스
    def __init__(self,kinds=""):
        self.kinds=kinds
    def getKinds(self):
        return self.kinds
class Pet: # 부모클래스
    def __init__(self, name=""):
        self.name=name
    def getName(self):
        return self.name
    
class Dog(Animal, Pet): # 다중상속
    def __init__(self, kinds="",name="",sound=""):
        Animal.__init__(self,kinds)
        Pet.__init__(self,name)
        self.sound=sound
    def getSound(self):
        return self.sound
    
class Cat(Animal, Pet):
    def __init__(self, kinds="", name="", age=()):
        Animal.__init__(self,kinds)
        Pet.__init__(self,name)
        self.age=age
    def getAge(self):
        return self.age
    
dog=Dog("개","퍼그","멍멍멍")

print("종류: "+dog.getKinds())
print("이름: "+dog.getName())
print("소리: "+dog.getSound())
print()

cat= Cat("고양이","먼치킨",30)
print("종류: "+cat.getKinds())
print("이름: "+cat.getName())
print("나이: "+str(cat.getAge()))
```

```python
# 추상메서드
# c++ 
# Java
# Python

# 추상메서드를 포함하면 추상 클래스가 된다.
# 추상 메서드는 구현하면 에러가 발생한다.
# 자식 클래스에서 반드시 오버라이드 해야 한다.
# 추상 클래스는 객체 생성이안된다.

class Pet:
    def __init__(self, name=""):
        self.name=name
    
    def dispName(self):
        print("이름 :"+ self.name)
        
class Dog(Pet):
    def __init__(self,name="",sound=""):
        Pet.__init__(self,name)
        self.sound=sound
    def dispSound(self):
        print("소리 :"+self.sound)
class Cat(Pet):
    def __init__(self, name="", size=0):
        Pet.__init__(self,name)
        self.size=size
    def dispSize(self):
        print("크기 :"+str(self.size))
    
dog=Dog("강아지","멍멍")
dog.dispName()
dog.dispSound()

cat=Cat("고양이",30)
cat.dispName()
cat.dispSize()
```

```python
import traceback
a=input("정수 :")
b=input("정수 :")
try :
    c=int(a)/ int(b)
    
except ValueError as e:
    print("정수만 입력하세요.")
    
except ZeroDivisionError as e:
    print("0으로 나누면 안됩니다.")
except:
    traceback.print_exc() # 에러난 경로를 쫓아가서 처리하라
# except Exception as e:
#     print(e)
else: # 예외가 없으면 아래를 출력해
    print(c)
finally:
    print("프로그램 끝")
    
    
```

```python
### 사용자 정의 예외
# 2~9 사이면 구구단 출력
# 아니면 잘못입력 출력
import traceback
class UserException(Exception) :
    def __init__(self, dan):
        print(str(dan)+":"+"2~9사이만 입력하세요.")
while True:
    dan= int(input("단 입력:"))
    try:
        if dan<2 or dan>9:
            raise UserException(dan) # 강제로 예외발생 throw
    except UserException as e:
        None
    
    except ValueError as e:
        print("정수를 입력하세요")
    except :
        traceback.print_exc()
    else:
        print(str(dan)+"단")
        for i in range(1,10):
            print(str(dan)+" * "+str(i)+" = "+str(dan*i))
        #구구단 출력
        break
    
```
